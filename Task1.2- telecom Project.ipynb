```python
def find_outlier(data1,threshold=1.5):
    outliers=pd.DataFrame()
    for i in data1.columns:
        if data1[i].dtype !='object':
            
            i_outlier=data1[(data1[i]<lowerrange)|(data1[i]>upperrange)]
            outlier=pd.DataFrame()
            outlier=pd.concat([outlier,i_outlier])
            data1[i]=np.where((data1[i]<lowerrange)|(data1[i]>upperrange),data1[i].mean(),data1[i])
    return outlier    
```
